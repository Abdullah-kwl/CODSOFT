#### Downsampling

```python
churn_majority=df[df['Exited']==0]
churn_minority=df[df['Exited']==1]
```

```python
from sklearn.utils import resample

churn_majority_downsample=resample(churn_majority,n_samples=2037,replace=False,random_state=42)
```

```python
churn_df=pd.concat([churn_majority_downsample,churn_minority])
```

```python
churn_df['Exited'].value_counts().to_frame()
```

```python
sns.countplot(data=churn_df,x='Exited')
plt.show()
```

```python
sns.heatmap(churn_df.corr(),annot=True)
plt.show()
```

```python
churn_df.columns
```

```python
churn_df.drop(['RowNumber', 'CustomerId', 'Surname','Geography','Gender'],axis=1,inplace=True)
```

```python
df_corr_exit=churn_df.corr()['Exited'].to_frame()
sns.barplot(data=df_corr_exit,x=df_corr_exit.index,y='Exited')
plt.xticks(rotation='vertical')
plt.show()
```

#### Spliting the Data Set

```python
x=churn_df.drop(['Exited'],axis=1)
y=churn_df['Exited']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

# Modeling and Evaluation

```python
from sklearn.metrics import confusion_matrix,recall_score,precision_score,accuracy_score,f1_score,ConfusionMatrixDisplay
```

```python
from sklearn.linear_model import LogisticRegression

lr=LogisticRegression(max_iter=500,n_jobs=-1)
lr.fit(x_train,y_train)
y_pred_lr=lr.predict(x_test)

print('Logistic Regression Report')
print('Accuracy Score =',accuracy_score(y_test,y_pred_lr))
print('Precision Score =',precision_score(y_test,y_pred_lr))
print('Recall Score =',recall_score(y_test,y_pred_lr))
print('F1 Score =',f1_score(y_test,y_pred_lr))
```

```python
cmd=ConfusionMatrixDisplay(confusion_matrix=confusion_matrix(y_test,y_pred_lr,labels=lr.classes_),display_labels=lr.classes_)
cmd.plot()
plt.title('Logistic Regression Confusion Matrix')
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier

rf=RandomForestClassifier(n_jobs=-1)
rf.fit(x_train,y_train)
y_pred_rf=rf.predict(x_test)

print('Random Forest Classifier Report')
print('Accuracy Score =',accuracy_score(y_test,y_pred_rf))
print('Precision Score =',precision_score(y_test,y_pred_rf))
print('Recall Score =',recall_score(y_test,y_pred_rf))
print('F1 Score =',f1_score(y_test,y_pred_rf))
```

```python
cmd=ConfusionMatrixDisplay(confusion_matrix=confusion_matrix(y_test,y_pred_rf,labels=lr.classes_),display_labels=lr.classes_)
cmd.plot()
plt.title('Random Forest Confusion Matrix')
plt.show()
```

```python
from sklearn.ensemble import GradientBoostingClassifier

gb = GradientBoostingClassifier(learning_rate=0.2, max_depth=5, n_estimators=300)
gb.fit(x_train,y_train)
y_pred_gb=gb.predict(x_test)

print('Gradient Boosting Classifier Report')
print('Accuracy Score =',accuracy_score(y_test,y_pred_gb))
print('Precision Score =',precision_score(y_test,y_pred_gb))
print('Recall Score =',recall_score(y_test,y_pred_gb))
print('F1 Score =',f1_score(y_test,y_pred_gb))
```

```python
cmd=ConfusionMatrixDisplay(confusion_matrix=confusion_matrix(y_test,y_pred_gb,labels=lr.classes_),display_labels=lr.classes_)
cmd.plot()
plt.title('Gradient Boosting Confusion Matrix')
plt.show()
```
